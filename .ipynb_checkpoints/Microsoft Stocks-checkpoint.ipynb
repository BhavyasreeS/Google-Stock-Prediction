### Dickey - Fuller Test

```python
# determine the rolling statistics (for Volume)
timeseries=data['Volume']
rolling_mean = timeseries.rolling(10).mean()
rolling_std = timeseries.rolling(10).std()
plt.figure(figsize=(20,10))
plt.plot(data.Volume, color="blue",label="Original Volume")
plt.plot(rolling_mean, color="red", label="Rolling Mean Volume")
plt.plot(rolling_std, color="black", label = "Rolling Standard Deviation in Volume")
plt.title("Microsoft Stock Time Series, Rolling Mean, Standard Deviation")
plt.legend(loc="best")
plt.show()
dftest = adfuller(timeseries, autolag='AIC')
dfout = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfout['Critical Value (%s)'%key] = value
print(dfout)
```

```python
# determine the rolling statistics (for Open value)
timeseries=data['Open']
rolling_mean = timeseries.rolling(10).mean()
rolling_std = timeseries.rolling(10).std()
plt.figure(figsize=(20,10))
plt.plot(data.Open, color="blue",label="Original Volume")
plt.plot(rolling_mean, color="red", label="Rolling Mean Volume")
plt.plot(rolling_std, color="black", label = "Rolling Standard Deviation in Volume")
plt.title("Microsoft Stock Time Series, Rolling Mean, Standard Deviation")
plt.legend(loc="best")
plt.show()
dftest = adfuller(timeseries, autolag='AIC')
dfout = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfout['Critical Value (%s)'%key] = value
print(dfout)
```

```python
# p-value > 0.05
# test-static > critical value, therefore we fail to reject the null hypothesis, i.e, H0 is accepted
# Also, the mean is increasing even though the std is small
# therefore the series is not stationary
```

#### Applying logarithmic transformation to detrend the series

```python
plt.figure(figsize=(20,10))
series_log=np.log(data['Volume'])
plt.plot(series_log)
plt.show()
```

```python
# Smoothing the series
plt.figure(figsize=(20,10))
moving_averge=series_log.rolling(window = 10).mean()
plt.plot(moving_averge , color='red')
```

```python
moving_avg_diff= series_log-moving_averge
moving_avg_diff
```

```python
moving_avg_diff.dropna(inplace=True)
moving_avg_diff
```

```python
timeseries=moving_avg_diff
rolling_mean = timeseries.rolling(10).mean()
rolling_std = timeseries.rolling(10).std()
plt.figure(figsize=(20,10))
plt.plot(timeseries, color="blue",label="Original")
plt.plot(rolling_mean, color="red", label="Rolling Mean ")
plt.plot(rolling_std, color="black", label = "Rolling Standard Deviation ")
plt.title("Microsoft Stock Time Series, Rolling Mean, Standard Deviation")
plt.legend(loc="best")
plt.show()
dftest = adfuller(timeseries, autolag='AIC')
dfout = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfout['Critical Value (%s)'%key] = value
print(dfout)
```

#### Removing seasonality using differencing

```python
plt.figure(figsize=(20,10))
series_log_diff=series_log-series_log.shift()
plt.plot(series_log_diff)
```

```python
series_log_diff.dropna(inplace=True)
timeseries=series_log_diff
rolling_mean = timeseries.rolling(10).mean()
rolling_std = timeseries.rolling(10).std()
plt.figure(figsize=(20,10))
plt.plot(timeseries, color="blue",label="Original")
plt.plot(rolling_mean, color="red", label="Rolling Mean ")
plt.plot(rolling_std, color="black", label = "Rolling Standard Deviation ")
plt.title("Microsoft Stock Time Series, Rolling Mean, Standard Deviation")
plt.legend(loc="best")
plt.show()
dftest = adfuller(timeseries, autolag='AIC')
dfout = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfout['Critical Value (%s)'%key] = value
print(dfout)
```

```python
# p-value < 0.05
# test-static < critical value, therefore we reject the null hypothesis, i.e, Ha is accepted
# Now the series is not stationary
```
